```python
# code 1 : Library catalogue system

class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))

    def display_catalog(self):
        print("Library Catalog:")
        for book in self.books:
            status = "Available" if book.available else "Not Available"
            print(f"{book.title} by {book.author} - {status}")

def main():
    library = Library()
    
    while True:
        print("\nLibrary Menu:")
        print("1. Add a new book")
        print("2. Display catalog")
        print("3. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            title = input("Enter book title: ")
            author = input("Enter author name: ")
            library.add_book(title, author)
        elif choice == "2":
            library.display_catalog()
        elif choice == "3":
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```

```python
class Book:
    def __init__(self, title, author, available=True):
        self.title = title
        self.author = author
        self.available = available

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, title, author):
        self.books.append(Book(title, author))

    def display_catalog(self):
        for book in self.books:
            status = "Available" if book.available else "Not Available"
            print(f"{book.title} by {book.author} - {status}")

    def borrow_book(self, title):
        for book in self.books:
            if book.title == title and book.available:
                book.available = False
                print(f"You have borrowed '{book.title}'.")
                return
        print("Book not found or not available.")

    def return_book(self, title):
        for book in self.books:
            if book.title == title and not book.available:
                book.available = True
                print(f"You have returned '{book.title}'.")
                return
        print("Book not found or already returned.")

def main():
    library = Library()
    while True:
        print("\nLibrary System Menu:")
        print("1. Add a new book")
        print("2. Display the catalog")
        print("3. Borrow a book")
        print("4. Return a book")
        print("5. Exit the program")

        choice = input("Enter your choice: ")
        
        if choice == "1":
            title = input("Enter the title of the book: ")
            author = input("Enter the author's name: ")
            library.add_book(title, author)
        elif choice == "2":
            library.display_catalog()
        elif choice == "3":
            title = input("Enter the title of the book you want to borrow: ")
            library.borrow_book(title)
        elif choice == "4":
            title = input("Enter the title of the book you want to return: ")
            library.return_book(title)
        elif choice == "5":
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please select a valid option.")

if __name__ == "__main__":
    main()
```

```python
# Code 2 : Matrix operations

def input_matrix(rows, cols):
    matrix = []
    for _ in range(rows):
        row = [float(x) for x in input().split()]
        matrix.append(row)
    return matrix

def print_matrix(matrix):
    for row in matrix:
        print(" ".join(str(cell) for cell in row))

def main():
    rows1, cols1 = map(int, input("Enter matrix 1 dimensions: ").split())
    matrix1 = input_matrix(rows1, cols1)

    rows2, cols2 = map(int, input("Enter matrix 2 dimensions: ").split())
    matrix2 = input_matrix(rows2, cols2)

    if rows1 != rows2 or cols1 != cols2:
        print("Matrix dimensions don't match for addition/multiplication.")
        return

    print("\nMatrix 1:")
    print_matrix(matrix1)

    print("\nMatrix 2:")
    print_matrix(matrix2)

    print("\nMatrix Addition:")
    result_addition = [[matrix1[i][j] + matrix2[i][j] for j in range(cols1)] for i in range(rows1)]
    print_matrix(result_addition)

if __name__ == "__main__":
    main()

```

```python
def input_matrix(rows, cols):
    matrix = []
    for i in range(rows):
        row = []
        for j in range(cols):
            element = float(input(f"Enter element at position ({i+1}, {j+1}): "))
            row.append(element)
        matrix.append(row)
    return matrix

def add_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix1[0])):
            row.append(matrix1[i][j] + matrix2[i][j])
        result.append(row)
    return result

def multiply_matrices(matrix1, matrix2):
    result = []
    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            cell = 0
            for k in range(len(matrix2)):
                cell += matrix1[i][k] * matrix2[k][j]
            row.append(cell)
        result.append(row)
    return result

def print_matrix(matrix):
    for row in matrix:
        print(row)

def main():
    rows1 = int(input("Enter the number of rows for matrix 1: "))
    cols1 = int(input("Enter the number of columns for matrix 1: "))
    matrix1 = input_matrix(rows1, cols1)

    rows2 = int(input("Enter the number of rows for matrix 2: "))
    cols2 = int(input("Enter the number of columns for matrix 2: "))
    matrix2 = input_matrix(rows2, cols2)

    print("\nMatrix 1:")
    print_matrix(matrix1)

    print("\nMatrix 2:")
    print_matrix(matrix2)

    if cols1 != rows2:
        print("Matrix dimensions are not suitable for multiplication.")
    else:
        print("\nMatrix Addition:")
        result_addition = add_matrices(matrix1, matrix2)
        print_matrix(result_addition)

        print("\nMatrix Multiplication:")
        result_multiplication = multiply_matrices(matrix1, matrix2)
        print_matrix(result_multiplication)

if __name__ == "__main__":
    main()
```

```python
# Code 3 : 

def find_combinations(nums, target):
    valid_combinations = []
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            for k in range(j + 1, len(nums)):
                if nums[i] + nums[j] + nums[k] == target:
                    valid_combinations.append((nums[i], nums[j], nums[k]))
    return valid_combinations

def main():
    nums = [int(x) for x in input("Enter numbers separated by spaces: ").split()]
    target = int(input("Enter the target sum: "))
    
    valid_combinations = find_combinations(nums, target)
    
    print(f"Unique combinations adding up to {target}:")
    for comb in valid_combinations:
        print(comb)

if __name__ == "__main__":
    main()
```

```python
from itertools import combinations

def find_combinations(nums, target):
    valid_combinations = []
    for comb in combinations(nums, 3):
        if sum(comb) == target:
            valid_combinations.append(comb)
    return valid_combinations

def main():
    nums = [int(x) for x in input("Enter a list of numbers separated by spaces: ").split()]
    target = int(input("Enter the target sum: "))
    
    valid_combinations = find_combinations(nums, target)
    
    print(f"Unique combinations adding up to {target}:")
    for comb in valid_combinations:
        print(comb)

if __name__ == "__main__":
    main()
```

```python
# code 4 : Sorting tuple

def input_tuple():
    name = input("Enter name: ")
    age = int(input("Enter age: "))
    height = float(input("Enter height: "))
    return name, age, height

def main():
    num_tuples = int(input("Enter the number of tuples: "))
    tuples = []
    
    for _ in range(num_tuples):
        tuple_data = input_tuple()
        tuples.append(tuple_data)
    
    tuples.sort()
    
    print("Sorted Tuples:")
    for tpl in tuples:
        print(tpl)

if __name__ == "__main__":
    main()
```

```python
# Code 5 : country capital

def main():
    country_capitals = {}
    lookup_attempts = 3
    
    while lookup_attempts > 0:
        country = input("Enter a country name: ").lower()
        capital = input(f"Enter the capital of {country.capitalize()}: ").lower()
        country_capitals[country] = capital
        
        lookup_attempts -= 1

    print("Country Capitals:")
    for country, capital in country_capitals.items():
        print(f"{country.capitalize()}: {capital.capitalize()}")

if __name__ == "__main__":
    main()
```

```python
import re
# re is for regular expression

def validate_name(name):
    return re.match("^[A-Za-z\s]+$", name)

def input_tuple():
    while True:
        name = input("Enter name: ")
        if validate_name(name):
            break
        else:
            print("Invalid name. Please enter a valid name (only alphabetic characters and spaces).")
    
    age = int(input("Enter age: "))
    height = float(input("Enter height: "))
    
    return name, age, height

def main():
    num_tuples = int(input("Enter the number of tuples: "))
    tuples = []
    
    for _ in range(num_tuples):
        tuple_data = input_tuple()
        tuples.append(tuple_data)
    
    tuples.sort(key=lambda x: (x[0], x[1], x[2]))
    
    print("Sorted Tuples:")
    for tpl in tuples:
        print(tpl)

if __name__ == "__main__":
    main()
```

```python
# Code 6 :  character frequency analysis

def analyze_character_frequency(text):
    frequency = {}
    for char in text.lower():
        if char.isalpha():
            frequency[char] = frequency.get(char, 0) + 1
    return frequency

def main():
    text = input("Enter a text: ")
    char_frequency = analyze_character_frequency(text)
    
    print("Character Frequency Analysis:")
    for char, count in char_frequency.items():
        print(f"'{char}': {count} times")

if __name__ == "__main__":
    main()
```
